Answer to the following Analytical questions: 

Question 1: Which country has gained the highest average IMdb Score according to our data? <br>
Use a bar chart to visualize your results

```python
country_avg_score = data.groupby('country')['imdb_score'].mean()
#country_avg_score.max()
#country_avg_score.idxmax()
country_avg_score.plot.bar(figsize =( 17,10))
```

Question 2: How many `colored` and `Black&White` movies do we have in our dataset?

```python
color_count = data[data.color.isin(['Color','Black and White'])]
color_count['color'].value_counts()
```

Question 3: Which movies of `Christopher Nolan` has the highest IMdb Score? <br>
Use a barchart to visualize the results

```python
nolan_movies = data[data.director_name=='Christopher Nolan']
#nolan_movies.shape
nolan_movies_scores = nolan_movies.set_index('movie_title')['imdb_score']
nolan_movies_scores.plot.bar(figsize =(9,6))

```

Question 4: Use a line chart to visualize how many movies have been produced in `USA` from `2010` to `2015`. 

```python
data.title_year = data.title_year.astype('Int64')
usa_movies = data[(data.country == 'USA') & (data.title_year.between(2010,2015))]
produc_movies = usa_movies.groupby('title_year').size()
final_visual =produc_movies.plot(color ='green', figsize =(10,6))
final_visual.set_xlabel('Years')
final_visual.set_ylabel('Number of Movies')
final_visual.set_title('Number of Movies Produced in USA (2010â€“2015)')
```
