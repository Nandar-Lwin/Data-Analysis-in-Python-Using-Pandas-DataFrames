### Let's answer to some analytical questions 
    
Question 1 - In which year we have the `highest` and `lowest` number of movies?

Now finding out number of movies per year

```python
count_per_year =data.groupby('title_year').size()
max_movie_counts = count_per_year.max()
year_max_movies = count_per_year.idxmax()
min_movie_counts =count_per_year.min()
year_min_movies = count_per_year.idxmin()
print(f"The year with the most movies is {year_max_movies} with {max_movie_counts}.")
print(f"The year with the fewest movies is {year_min_movies} with {min_movie_counts}.")
```

Let's create a line chart to show the result

```python
ax =count_per_year.plot(figsize=(8,6))
ax.set_xlabel('Years')
ax.set_ylabel('Counts of Movie')
```

Question 2 - In which year we have `minimum` and `maximum` average `imdb_score` ? 

Finding average imdb score per year 

```python
avgscores = data.groupby('title_year')['imdb_score'].mean()
max_avg_score = avgscores.max()
year_max_score = avgscores.idxmax()
min_avg_score = avgscores.min()
year_min_score = avgscores.idxmin()
print(f"The year with the most average score is {year_max_score} with {max_avg_score}.")
print(f"The year with the fewest average score is {year_min_score} with {min_avg_score}.")
```

Let's Use a barchart to visualize the result

```python
avg_bar = avgscores.plot.bar(figsize =(18,8) ,color ='brown')
avg_bar.set_title('Average IMDb Score per year')
avg_bar.set_xlabel('Years')
avg_bar.set_ylabel('Avg Score')
```
